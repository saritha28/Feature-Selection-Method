```python
df.drop("STUDENT ID",axis=1, inplace=True)
```

```python
x=df.drop("GRADE",axis=1)
```

```python
y=df.GRADE
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
chi_select=SelectKBest(chi2,k=15)
```

```python
chi_select.fit(x,y)
```

```python
chi_select.get_support()
```

```python
feat=chi_select.get_support()
```

```python
x.loc[:,feat].columns.tolist()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
rfe=RFE(LogisticRegression(),n_features_to_select=15)
```

```python
rfe.fit(x,y)
```

```python
rfe.ranking_
```

```python
for i in zip(df.columns,rfe.support_,rfe.ranking_):
    print(i)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
Select=SelectFromModel(RandomForestClassifier(),max_features=15)
```

```python
Select.fit(x,y)
```

```python
col=Select.get_support()
```

```python
x.loc[:,col].columns.tolist
```
